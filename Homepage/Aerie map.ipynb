```python
# the dataframe library
import pandas

# read in the data with questions as columns and people as rows as each row in the appropriate CSV format (note the filename)
people = pandas.read_csv('aeriedata_attending.csv', sep=',',header=0, index_col = 'Q2')
# we don't need the timestamps for analysis
#del people['Timestamp']

print("there are",people.shape[0],"people and", people.shape[1],"questions")
# show the first five
people.head()
```

```python
# note, many of the analyses below we will perform for understanding
# individuals, but that same can be done to analyze the questions too!

# create the row x column matrix of questions x people responses
questions = people.transpose()
questions.head(100)
```

```python
# make a vector from each person
people['data'] = people['Q4'] + ',' + people['Q3'] + ',' + people['Q7'] + ',' + \
    people['Q9.1'] + ',' + people['Q8'] + ',' + people['Q10'] + ',' + people['Q11']
people['data']
people_onehot = people['data'].str.get_dummies(sep=",")

questions_onehot = people_onehot.transpose() #used later

people_onehot
```

```python
# find features which correlate to create a simple model for maximum prediction
corr_feature = 'Artificial Intelligence'
# the data frame which holds correlations, sorted from highest to lowest
corr_df = people_onehot.corr().sort_values(corr_feature,ascending=False)[corr_feature]

print("The highest 5 correlated features:\n",corr_df.head(),"\n")
print("The lowest 5 (most negatively correlated) features\n",corr_df.tail())
```

```python
# let's do a cluster analysis of questions
from sklearn import cluster
k = 10
k_means = cluster.KMeans(n_clusters=k)
k_means.fit(questions_onehot.values)

cluster_labels = k_means.labels_
#print(k_means.labels_)
#print(list(questions.index))
#questions.index[cluster_labels == 2]
for i in range(k):
    print (i, list(questions_onehot.index[cluster_labels == i]))
```

```python
# let's do a cluster analysis of people
k = 10
k_means = cluster.KMeans(n_clusters=k)
k_means.fit(people_onehot.values)

cluster_labels = k_means.labels_
people['aerie'] = cluster_labels
#print(k_means.labels_)
#print(list(people.index))
#people.index[cluster_labels == 2]
for i in range(k):
    print (i, list(people_onehot.index[cluster_labels == i]))

```

```python
# instead of cluster, let's put this all on a graph
# but how do we reduce so many numbers down to two? (for a 2D plot)
# dimensionality reduction!

from sklearn import decomposition
pca = decomposition.PCA(n_components=6)
pca.fit(people_onehot)

X = pca.transform(people_onehot)

print(X.shape)

import pylab as pl
import numpy as np
pl.scatter(X[:, 0], X[:, 1], c=cluster_labels, s=50)

# a plot without labels
pl.show()
```

```python
pl.figure(figsize=(10,8), dpi=300)
pl.scatter(X[:, 0], X[:, 1], c=cluster_labels, s=50)
for label, x, y in zip(list(people_onehot.index), X[:, 0], X[:, 1]):
    pl.annotate(
        label, 
        xy = (x, y), xytext = (30*x,30*y),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))

# a plot with labels
pl.show()
```

```python
# understanding the components
# first, how much of the variation in the data does this graph represent?

#print(list(questions.index))
#print(pca.components_)

#suppress the default printing of float point notation for numbers
np.set_printoptions(suppress=True)

print('variance explained by these components:',pca.explained_variance_ratio_)
```
